```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv("Lollapalooza_Argentina_2025_Dataset.csv")

# Mostrar las primeras filas
df.head()
```

```python
# Información general del DataFrame
df.info()
```

```python
# Tamaño del dataset
print("Filas y columnas:", df.shape)

# Tipos de datos y nulls
print("\nInformación general del dataset:")
df.info()

# Ver si hay valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
# Estadísticas descriptivas generales
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gráfico de dispersión
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Asistentes estimados', y='Popularidad (Spotify mensual)')
plt.title('Asistentes estimados vs Popularidad')
plt.xlabel('Asistentes estimados')
plt.ylabel('Oyentes mensuales en Spotify')
plt.grid(True)
plt.show()

# Gráfico de barras por artista
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='Artista', y='Asistentes estimados')
plt.title('Asistentes por artista en Lollapalooza 2025')
plt.xticks(rotation=45)
plt.ylabel('Asistentes estimados')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Género', y='Asistentes estimados', estimator=sum, ci=None)
plt.title('Asistentes por Género Musical en Lollapalooza 2025')
plt.ylabel('Total Asistentes Estimados')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='Día', y='Asistentes estimados', estimator=sum, ci=None)
plt.title('Asistentes por Día del Festival')
plt.ylabel('Total Asistentes Estimados')
plt.tight_layout()
plt.show()
```

```python
top_spotify = df.sort_values(by='Popularidad (Spotify mensual)', ascending=False).head(3)
plt.figure(figsize=(8,5))
sns.barplot(data=top_spotify, x='Artista', y='Popularidad (Spotify mensual)', palette='crest')
plt.title('Top 3 Artistas más Populares en Spotify')
plt.ylabel('Oyentes mensuales')
plt.tight_layout()
plt.show()
```
